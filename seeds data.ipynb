```python
import pandas as pd
pd.set_option('display.max_columns', 500)

# import galapagos_potato as bp
import back_propagation as bp
import quesadilla as qd
```

```python
temp = pd.read_csv(r'C:\Users\Chulabutrach\Documents\Coding\Projects\pttgc\neural_network\data\Seeds\seeds_only_1_2.csv')
# df['label'] = df['fetal_health'].apply(lambda x: 0 if x == 1 else 1)
df = qd.normalization(temp, ['Area', 'Perimeter', 'Compactness', 'Kernel.Length', 'Kernel.Width', 'Asymmetry.Coeff', 'Kernel.Groove'])
df.head(100)
```

```python
df.columns
```

```python
df['Label'] = df['Type'].apply(lambda x: 0 if x == 2 else 1)
# print(df.tail())
# df_good = df.loc[df['Label'] == 1]
# df_bad = df.loc[df['Label'] == 0]

# df_bad = df_bad.sample(n=len(df_good), random_state=10)
# df = df_good.append(df_bad).reset_index(drop=True)
# df = df.sample(frac=1).reset_index(drop=True)
# df = df.reset_index()
# df.groupby('Label').index.nunique().plot.bar()
# df = df.drop(['index', 'Type'], axis=1)

df = df.sample(frac=1).reset_index(drop=True)
df = df.drop(['Type'], axis=1)

df_test = df.tail(12)
df = df.head(120)
# df = df.sort_values('Label')
df.head()
```

```python
df.shape
```

```python
x = df
y = x.pop('Label').values.tolist()
x = x.values.tolist()
```

```python
n_inputs = len(x[0])
print(n_inputs)
n_outputs = 2
network = bp.initialize_networks(n_inputs, 5, n_outputs, n_layers=1)
bp.train_network(network, x, y, 0.15, 300, n_outputs)
```

```python
x_test = df_test
y_test = x_test.pop('Label').values.tolist()
x_test = x_test.values.tolist()
```

```python
accuracy = bp.accuracy_evaluation(network, x_test, y_test)
print(accuracy)
```
